# MatPlotLib Overview

### Introduction

- MatPlotLib is the "grandfather" library of datavisualization with Python.
- created by: John Hunter
- He created it to replicate Matlib's plotting capabilities in Python.
- it's an excellent 2D and 3D graphics library for generating scientific figures.

#### Some Major Pros:
- Generally easy to get started for simple plots.
- Support for custom lebels and texts
- Great control of every element in a figure
- High-quality output in many formats
- Very customizable in general

```python
import matplotlib.pyplot as plt
```

- 2 methods of plotting MatPlotLib plot:
    1. Functional Method
    2. Object Oriented Method (better)

### 1. Functional Method

```python
plt.plot(x,y,'g*-',linewidth=2,)
# plt.plot(x,y,'b--',linewidth=2,)
# plt.plot(x,y,'r.-',linewidth=2,)
# plt.plot(x,y,'r-*',linewidth=2,)
# plt.plot(x,y,'r*',linewidth=2,)
# plt.plot(x,y,'o-',linewidth=2,)

plt.title("Square Numbers", fontsize=18)
plt.xlabel("Data", fontsize=15)
plt.ylabel('Square of Data', fontsize=15)
plt.tick_params(axis='both',labelsize=10)
```

```python
plt.scatter(x,y,c='red',edgecolors='none',s=30)

plt.title("Square Values Plot", fontsize=18)
plt.xlabel("Data -->", fontsize=15)
plt.ylabel("Square of Data -->", fontsize=15)
plt.tick_params(axis="both", labelsize=10)
```

```python
plt.scatter(x,y,c=y,cmap=plt.cm.Blues,edgecolors='none',s=30)
plt.title("Squaring Plot", fontsize=18)
plt.xlabel("Data -->", fontsize=15)
plt.ylabel("Square -->", fontsize=15)
plt.tick_params(axis="both",labelsize=10)
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'o-')

plt.subplot(1,2,2)
plt.plot(y,x,'g*-')
```

### 2. Object-Oriented Method:
formal Introduction of MatPlotLib's Object-Oriented API.

#### - Introduction:
the main idea in using the formal Object Oriented method is to create a figure objects and then just call methods or attributes of that object.
   - this approach is much better when dealing with the canvas that has multiple plots.

```python
# create figure: empty canvas
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left, buttom, width, height (range 0-1)

# Plot on that set of axes
axes.plot(x,y,'g*-',linewidth=1)
axes.set_xlabel("X label", fontsize=15)
axes.set_ylabel("Y label", fontsize=15)
axes.set_title("Square Plot", fontsize=18)
axes.tick_params(axis='both',labelsize=10)
```

```python
# create blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # main axes
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # inset axes
# axes2 = fig.add_axes([0.5,0.15,0.4,0.3])

axes1.plot(x,y, 'go-', linewidth=2)
axes1.set_title("Larger Canvas")

axes2.plot(y,x,'g*-', linewidth=1)
axes2.set_title("Smaller Canvas")
```

### Subplots:
"plt.subplots()" object will act as a more automatic axis manager.

```python
fig, axes = plt.subplots()

axes.plot(x,y,'r')
axes.set_xlabel("X label --->")
axes.set_ylabel("Y label --->")
axes.set_title("Canvas")
```

then we can specify no of rows and columns to subplots.

```python
# an empty canvas with 2 plots
fig, axes = plt.subplots(nrows=1,ncols=2)

# fig, axes = plt.subplots(nrows=3, ncols=3)
# plt.tight_layout()
```

```python
# axes is an array of Axes to plot on.
axes
```

```python
for ax in axes:
    ax.plot(x,y,'r')
    ax.set_title("Plot")
    ax.set_xlabel("X-label --->")
    ax.set_ylabel("Y-label --->")
# Display the figure object
fig
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y,'r')
axes[0].set_title("Plot-1")
axes[0].set_xlabel("X-label --->", fontsize=10)
axes[0].set_ylabel("Y-label --->", fontsize=10)


axes[1].plot(y,x,'b')
axes[1].set_title("Plot-2")
axes[1].set_xlabel("X-label --->", fontsize=10)
axes[1].set_ylabel("Y-label --->", fontsize=10)

fig.tight_layout()
```

### Figure Size, Aspect Ratio and DPI

Matplotlib allows to add these figure properties in plot.
- figsize : tuple of width and height of figure in inches.
- dpi : dots-per-inches (pixels-per-inches)

```python
fig = plt.figure(figsize=(6,2), dpi=100)
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
```

The same arguments can also be sent to Layout Managers, such as "subplots()" function

```python
fig, axes = plt.subplots(nrows=1,ncols=2,figsize=(8,4))
axes[0].plot(x,y,'r')
axes[0].set_title("Plot-1",fontsize=12)
axes[0].set_xlabel("x-label -->",fontsize=10)
axes[0].set_ylabel("y-label -->",fontsize=10)

axes[1].plot(y,x,'g')
axes[1].set_title("Plot-2",fontsize=12)
axes[1].set_xlabel("x-label -->",fontsize=10)
axes[1].set_ylabel("y-label -->",fontsize=10)

fig.tight_layout()
```
