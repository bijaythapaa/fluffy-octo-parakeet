## Setting Colors, Linewidths, Linetypes

### Colors with Matlab like syntax
(avoid using)

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
fig,ax=plt.subplots()
ax.plot(x,x**2,'b.-')
ax.plot(x,x**3,'g--')
```

### Colors with color=parameter
- optional argument 'alpha=0.5' : range(0-1), indicates Opacity

```python
fig, axes = plt.subplots()

axes.plot(x, x+1, color="blue", alpha=0.5) # half-transparant
axes.plot(x, x+2, color="#8B008B")        # RGB hex code
axes.plot(x, x+3, color="#FF8C00")        # RGB hex code 
```

### Line and Marker Style

```python
fig,ax = plt.subplots(figsize=(12,6))

ax.plot(x,x+1,color='red',linewidth='0.25')
ax.plot(x,x+2,color='red',linewidth='0.50')
ax.plot(x,x+3,color='red',linewidth='1.00')
ax.plot(x,x+4,color='red',linewidth='2.00')

# possible linestyle options '-','--','-.',':','solid','dashdot','dotted',' ','','dashed'
ax.plot(x,x+5,color='green',lw='3',linestyle='-')
ax.plot(x,x+6,color='green',lw='3',ls='-.')
ax.plot(x,x+7,color='green',lw='3',ls=':')

# custom dash
line,=ax.plot(x,x+8,color='black',lw=1.50)
line.set_dashes([5,10,15,10])

# possible marker symbols: marker = '+','o','*','s','.','1','2','3','4',...
ax.plot(x,x+9,color='blue',lw=3,ls='-',marker='+')
ax.plot(x,x+10,color='blue',lw=3,ls='--',marker='o')
ax.plot(x,x+11,color='blue',lw=3,ls=':',marker='s')
ax.plot(x,x+12,color='blue',lw=3,ls='--',marker='1', markersize=5)

# marker size and color
ax.plot(x,x+13,color='purple',lw=1,ls='-',marker='o',markersize=2)
ax.plot(x,x+13,color='purple',lw=1,ls='-',marker='o',markersize=4)
ax.plot(x,x+13,color='purple',lw=1,ls='-',marker='o',markersize=8, markerfacecolor="red")
ax.plot(x,x+13,color='purple',lw=1,ls='-',marker='s',markersize=9,
       markerfacecolor='yellow',markeredgewidth=3,markeredgecolor='green')
```

### Control Over Axis Apperance
    this section looks at controlling axis sizing properties in a matplotlib figure.

### Plot Range
we can configure the ranges of axes using the 'set_xlim' and 'set_ylim' methods in axis object, or '.axis('tight') for automatically getting "tightly fitted" axes ranges.

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color="purple",lw=2,ls="--",)
ax.set_xlim([0,1])
ax.set_ylim([0,2])
# ax.set_xlim([1,5])
# ax.set_ylim([5,35])
```

```python
fig, axes = plt.subplots(1,3,figsize=(12,4))

axes[0].plot(x,x**2,x,x**3)
axes[0].set_title('default axes range')

axes[1].plot(x,x**2,x,x**3)
axes[1].axis('tight')
axes[1].set_title('tight axes')

axes[2].plot(x,x**2,x,x**3)
axes[2].set_ylim([0,60])
axes[2].set_xlim([2,5])
axes[2].set_title('custom axes range')
```
