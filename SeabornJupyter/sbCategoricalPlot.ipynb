## Categorical Plot
lets discuss using seaborn to plot categorical data. Few main plot types:
- factorplot
- boxplot
- voilinplot
- stripplot
- swarmplot
- barplot
- countplot

let's go through example of each.

```python
import seaborn as sns
%matplotlib inline
```

```python
import numpy as np
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

## Barplot and Countplot
These are very similar plots allows us to get aggregate data off a categorical feature in our data.

### Barplot:
a general plot that allows us to aggregate the categorical data based off some function, by default the 'mean'.
like: visualization of a group-by action.

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

### countplot
essentially same as barplot except the "estimator" is explicitly counting the no. of occurences. which is why we only pass the x-values.

```python
sns.countplot(x='sex',data=tips)
```

## BoxPlot and ViolinPlot
boxplot and violin plots are used to show the distribution of categorical data.

### Boxplot
shows distribution of quantitative data in a way that facilates comparisions between vaiables or acress levels of a categorical variable.
- the 'box' shows quartile of dataset while 'whiskers' extend to show the rest of the distribution, except for points that are 'outliers' using a method i.e. a function of the inter-quartile range.

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker',palette='coolwarm')
```

```python
# can do entire dataframe with orient 'h'
sns.boxplot(data=tips,palette='rainbow',orient='h')
```

### violinplot
plays similar role as a boxplot. shows the distribution of quantitative data across several levels of one (or more) categorical variable such as that those distributions can be compared.
- unlike a boxplot, in which all of the plot components corresponds to actual datapoints, the violin plot features a kernal density estimation(KDE) of underlying funtion. (allows us to reveal more info about data points themselves but little hard to interpret) 

```python
sns.violinplot(x='day',y='total_bill',data=tips,palette='rainbow')
```

```python
sns.violinplot(x='day',y='tip',data=tips,hue='sex',palette='Set1')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True,palette='Set1')
```

```python
sns.violinplot(x='day',y='tip',hue='time',data=tips,split=True,palette='Set1')
```

## Stripplot and Swarmplot

### stripplot
draws a scatter plot where one variable is categorical(another is numerical). 
A stripplot can be drawn on its own, but its good complement to box or violin plot in cases where you want to show all observations along with some representation of the underlying distribution.
- drawback: we cant tell how many data-points are stacked on top of each-other.(for good, we can pass argument 'jitter=True', but still not enough)

```python
sns.stripplot(x='day',y='total_bill',data=tips)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',palette='Set1')
```

```python
# sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',palette='Set1',split=True)
# parameter 'split' is renamed to 'dodge'
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',palette='Set1',dodge=True)
```

### swarmplot
similar to stripplot but the points are adjusted (only along categorical axis) so that they dont overlap. 
This gives better representation of distribution of values, although it doesnot scale very well to large numbers of observations both in terms of the ability to show all the points and in terms of the computation needed to arrange them.
- do not use for very large data-set 
- combines idea of a stripplot and violinplot

```python
sns.swarmplot(x='day',y='total_bill',data=tips)
```

```python
# sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex',palette='Set1',split=True)
# parameter 'split' is renamed to 'dodge'
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex',palette='Set1',dodge=True)
```

## combining categorical plot
useful for exploratory data analysis. not for data-visualization.

```python
sns.violinplot(x='tip',y='day',data=tips,palette='rainbow')
sns.swarmplot(x='tip',y='day',data=tips,color='black',size=3)
```

## CatPlot
the most general form of categorical plot. it can take-in a 'kind' operator to adjust the plot type.
- renamed 'catplot' in newer version, old name was 'factorplot'

```python
# sns.factorplot(x='day',y='total_bill',data=tips,kind='bar')
sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
```

```python
sns.catplot(x='sex',y='total_bill',data=tips,kind='violin')
```
