## Regression Plots
Seaborn has many built-in capabilities for regression plot.however we won't really discuss regression until the machine learning section of the course, so we will only cover the lmplot() function for now.

#### Lmplots:
allows us to display linear models, but it conveniently allows us to split up those plots based off of features, as well as coloring the hue based off of features.

```python
import seaborn as sns
# import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

#### lmplot()

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',hue='sex',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',hue='sex',data=tips,markers=['o','v'],palette='Set1')
```

#### Working with Markers
Implots kwargs get pass through to 'regplot' which is more general form of lmplot(). 'regplot' has a 'scatter_kws' parameter that gets passed to "plt.scatter".
So, we want to set the 's' parameter in that dictionar, which corresponds to squared markersize.
In other word, we end up passing a dictionary with the base matplotlib arguments, in this case, 's' for size of a scatter plot.
reference: https://matplotlib.org/api/markers_api.html

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='Set1',
           markers=['o','v'],
           scatter_kws={'s':70},)
```

### Using a Grid
we can add more variable separation through columns and rows with the use of grid.
Just indicate it with the 'col' or 'row' arguments.
(we can say, use grid instead of hue)

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',hue='sex',palette='coolwarm').add_legend()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,row='time',hue='sex',palette='coolwarm').add_legend()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='day',row='time',hue='sex',palette='Set1').add_legend()
```

#### Aspect and Size
Seaborn figures can have their size and aspect ratio adjusted with "size" and "aspect" parameters.

```python
sns.lmplot(x='total_bill',y='tip',data=tips,
          col='day', hue='sex',
          aspect=0.6, size=8,)
# size parameter is renamed to height in newer version.
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,
          col='day',hue='sex',
          aspect=0.6, height=8)
```
