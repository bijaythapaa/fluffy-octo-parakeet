## Grids
Grids are general type of plots that allows us to map plot types to row and colomns of a grid, this helps us creating similar plots seperated by features.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

### 1. PairGrid
a Subplot grid for plotting pairwise relationship in dataset.

```python
sns.PairGrid(iris)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diagonal(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(plt.kdeplot)
```

#### A. Pairplot:
 - a similar version of PairGrid (we'll often use)

```python
sns.pairplot(data=iris,hue='species',palette='rainbow')
```

```python
sns.pairplot(iris)
```

### 2. Facet Grid
a general way to create grids of plots based off a fature.

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
g = sns.FacetGrid(data=tips,row='time',col='smoker')
```

```python
g = sns.FacetGrid(tips,row='time',col='smoker')
# g.map(sns.distplot,'total_bill')
g.map(sns.displot,'total_bill')
```

```python
g = sns.FacetGrid(tips,row='smoker',col='time')
g = g.map(plt.hist,'total_bill')
```

```python
g = sns.FacetGrid(tips,row='time',col='smoker')
g = g.map(plt.scatter,'total_bill','tip')
```

```python
g=sns.FacetGrid(tips,row='smoker',col='time',hue='sex')
g=g.map(plt.scatter,'total_bill','tip').add_legend()
```

### 3. Joint Grid
general version for jointplot() type grids.

```python
g = sns.JointGrid(x='total_bill',y='tip',data=tips)
```

```python
g = sns.JointGrid(x='total_bill',y='tip',data=tips)
# g = g.plot(sns.regplot,sns.displot)
g = g.plot(sns.regplot,sns.histplot)
```
