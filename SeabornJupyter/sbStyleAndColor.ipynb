# Style and Color
we've seen how to control figure aesthetics in seaborn.

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips = sns.load_dataset('tips')
tips.head()
```

## Styles
we can set particular styles.

```python
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('white')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
```

### Spine Removal

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
sns.despine()
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
sns.despine(left=True,bottom=True)
```

```python
sns.set_style('ticks')
sns.countplot(x='sex',data=tips)
sns.despine(right=False, bottom=True)
```

## Size and Aspect:
we can use matplotlib: 'plt.figure(figsize=(width,height))' to change size of most seaborn figure. 
we can change size and aspect ratio of most seaborn grid plots by passing in parameter: size and aspect.

```python
# Non grid plot
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

```python
# Grid Type plot
sns.lmplot(x='total_bill',y='tip',height=2,aspect=3,data=tips)
```

### Scale and Context
"set_context()": allows us to override default parameter. (and change font-scale depending on what we're actually creating the figure for.)

```python
sns.set_context('poster',font_scale=0.8)
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('talk',font_scale=1)
sns.countplot(x='sex',data=tips)
```

```python
sns.set_context('talk',font_scale=1,rc={'lines.linewidth':5})
sns.countplot(x='sex',data=tips)
```

### Palette and Color:
see: matplotlibs official docs colormap page. there's huge list of colormap names.
Eg. viridis, inferno, plasma, magma, set1,set2,set3,seismic, coolwarm,...

```python
# sns.set_context()
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',palette='inferno')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,height=5,aspect=1.5,hue='sex',palette='magma')
```
